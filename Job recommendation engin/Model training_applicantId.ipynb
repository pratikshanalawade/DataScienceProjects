## Creating the Model

```python
# importing the libraries

import nltk
from nltk.stem.porter import PorterStemmer 
from nltk.corpus import stopwords
import re
```

```python
ps = PorterStemmer()
```

```python
# creating a function that will be cleaning our text
# tokenize and remove unnecessery keywords in out text

def cleaning(text):
    clean_text = re.sub(r'[^a-zA-Z0-9\s]','',text)
    tokenize = nltk.word_tokenize(clean_text.lower())
    stemming = [ps.stem(words) for words in tokenize if words not in stopwords.words('english')]
    return " ".join(stemming)
```

```python
cleaning("left and moving loving //85684 @#$%^&")
```

```python
# creating a copy of new_data
df1 = new_data.copy()

# applying text cleaning for each column
for i in df1.columns:
    if df1[i].dtype != "O":
        df1[i] = df1[i]
    else:
        # using lambda to iterate for each row
        df1[i] = df1[i].apply(lambda x: cleaning(x))
# returning the data_set
df1
```

```python
#combining all columns
df1['new_col'] = ''
for i in df1.columns[2:-1]: # excluding new column
    df1['new_col'] = df1['new_col']+ " " + df1[i] 
    
# df1['new_col'] = df1.apply(lambda row: ''.join(row.values.astype(str)), axis=1)
```

```python
df1.head(2)
```

```python
# importing the TFIDF and COSINE SIMILARITY METRICES

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfif = TfidfVectorizer() 
matrix = tfif.fit_transform(df1['new_col'])
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python
def recommendation(title):
    # Check if the title exists in the DataFrame
    if title not in df1['applicantId'].values and title not in df1['jobId'].values:
        return []

    # Initialize lists to store distance indices
    distance_idx = []
    distance_idx_2 = []

   
    # Find the index of the given job title.
    if title in df1['jobId'].values:
        idx_list = df1[df1['jobId'] == title].index
        idx = df1.index.get_loc(idx_list[0])
        # Calculate similarity indices and sort them
        distance_idx = sorted(list(enumerate(similarity[idx])), key=lambda x: x[1], reverse=True)[1:5]
    """
    he enumerate function adds a counter to each item in the iterable (in this case, similarity[idx]). 
    This means that enumerate(similarity[idx]) will create an iterable of tuples, 
    where each tuple contains an index and the corresponding similarity score.
    """
    # Find the index of the given current designation.
    if title in df1['applicantId'].values:
        idx_2_list = df1[df1['applicantId'] == title].index
        idx_2 = df1.index.get_loc(idx_2_list[0])
        # Calculate similarity indices and sort them
        distance_idx_2 = sorted(list(enumerate(similarity[idx_2])), key=lambda x: x[1], reverse=True)[1:5]

    # Retrieve the job titles for the top similar jobs, avoiding duplicates
    jobs = []
    seen = set()
    for i in distance_idx:
        job_title = df1.iloc[i[0]].jobTitle
        if job_title not in seen:
            jobs.append(job_title)
            seen.add(job_title)
    for i in distance_idx_2:
        job_title = df1.iloc[i[0]].currentDesignation
        if job_title not in seen:
            jobs.append(job_title)
            seen.add(job_title)

    return jobs
```

```python
# for applicant id
recommendation("aisb2256")
```

```python
# for Job ID
recommendation("jdsd2289")
```
